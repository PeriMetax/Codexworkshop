# Demo notebook for testing\n\nThis notebook creates dummy data so you can test workflows without any external files.

```python
import random\nfrom datetime import date, timedelta\n\nrandom.seed(42)
```

```python
# Generate dummy transaction-like data\ncategories = ['Books', 'Electronics', 'Home', 'Grocery', 'Toys']\nstart = date(2026, 1, 1)\nrows = []\nfor i in range(100):\n    rows.append({\n        'id': i + 1,\n        'date': (start + timedelta(days=i % 30)).isoformat(),\n        'category': random.choice(categories),\n        'amount_usd': round(random.uniform(5, 500), 2),\n        'is_returning_customer': random.choice([True, False]),\n    })\n\nrows[:3]
```

```python
# Optional: convert to pandas DataFrame (if pandas is installed)\ntry:\n    import pandas as pd\n    df = pd.DataFrame(rows)\n    display(df.head())\nexcept ModuleNotFoundError:\n    print('pandas is not installed; using raw Python data structures.')
```

```python
# Basic summary that works with or without pandas\nif 'df' in globals():\n    print('Total rows:', len(df))\n    print('Average amount:', round(df['amount_usd'].mean(), 2))\n    print('By category:')\n    display(df.groupby('category')['amount_usd'].agg(['count', 'mean']).round(2))\nelse:\n    total = len(rows)\n    avg = sum(r['amount_usd'] for r in rows) / total\n    print('Total rows:', total)\n    print('Average amount:', round(avg, 2))
```

```python
# Save dummy data to CSV if pandas is available\nif 'df' in globals():\n    out_path = 'demo/dummy_data.csv'\n    df.to_csv(out_path, index=False)\n    print(f'Saved: {out_path}')\nelse:\n    print('Skipping CSV export because pandas is not installed.')
```
